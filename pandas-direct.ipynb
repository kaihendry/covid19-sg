```python
import pandas as pd
import glob
```

```python
all_files = glob.glob("/home/hendry/tmp/COVID-19/csse_covid_19_data/csse_covid_19_daily_reports/*.csv")

li = []
# https://github.com/simonw/covid-19-datasette/blob/4ce3738229048f235571f93310c686c00cbf7c6b/build_database.py#L19
for filename in all_files:
    df = pd.read_csv(filename, index_col=False)
    df.rename(columns={'Country/Region':'Country_Region'}, inplace=True)
    df.rename(columns={'Last Update':'Last_Update'}, inplace=True)
    df = df[(df['Country_Region'] == "Singapore")]
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame['Last_Update'] = pd.to_datetime(frame['Last_Update'], infer_datetime_format=True)
```
