# Running total to daily count

https://stackoverflow.com/questions/63520137/turning-a-running-total-to-a-daily-count/

```gnuplot
reset session
set datafile separator comma
set format x "%m/%d" timedate
x1=NaN
y1=NaN
plot "singapore.csv" u (x0=x1,x1=timecolumn(2,"%Y-%m-%d")):(y0=y1,y1=$5,(y1-y0)/(x1-x0)*3600*24) \
    w lp pt 7 title "Daily change"
```

# From August 2020-08-01

This is how [zoom or range into a particular time period](https://stackoverflow.com/questions/62993755/gnuplot-time-range-limit)

```gnuplot
StartTime = strptime("%Y-%m-%d","2020-08-01")
set xrange[StartTime:]
plot "singapore.csv" u (x0=x1,x1=timecolumn(2,"%Y-%m-%d")):(y0=y1,y1=$5,(y1-y0)/(x1-x0)*3600*24) \
    w lp pt 7 title "Daily change"
```
